```python
# Histogram for math score in the dataset
# Continuous variable

plt.figure(figsize=(8,5))
plt.hist(df["math_score"], bins=4,color='blue',edgecolor='red')
plt.xlabel("score")
plt.ylabel("Average number of students")
plt.show()
```

```python
#Bar chart for Gender in the dataset
# Categorical variable

plt.figure(figsize = (8,5))
df['gender'].value_counts().plot(kind='bar', color =['red','purple'],edgecolor='black')
plt.xlabel("gender")
plt.ylabel("Frequency")
plt.show()
```

```python
#Bar chart for extracurricular_activities in the dataset
# Categorical variable

plt.figure(figsize = (7,5))
df['extracurricular_activities'].value_counts().plot(kind='bar', color =['blue','lightgreen','yellow'],edgecolor='black')
plt.xlabel("extracurricular_activities")
plt.ylabel("Frequency")
plt.show()
```

```python
#Bar chart for part_time_job in the dataset
# Categorical variable

plt.figure(figsize = (8,5))
df['part_time_job'].value_counts().plot(kind='bar', color =['orange','brown'],edgecolor='black')
plt.xlabel("part_time_job")
plt.ylabel("Frequency")
plt.show()
```

```python
#Histogram for study time weekly of the students in the dataset
# Continuous variable

plt.figure(figsize=(8,5))
plt.hist(df["weekly_self_study_hours"], bins=100,color='yellow',edgecolor='black')
plt.xlabel("Study time weekly of the Students")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram for absence_days of the Students in the dataset
# Continuous variable

plt.figure(figsize=(10,5))
plt.hist(df["absence_days"], bins=30,color='pink',edgecolor='black')
plt.xlabel("absence_days of the Students")
plt.ylabel("Frequency")
plt.show()
```
